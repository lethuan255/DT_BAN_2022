```python
def k_fold(k, data_frame, name_prefix, root_path):
  start = 0
  step = (int)(data_frame.shape[0] / k)
  end = 0
  for i in range (0, k):
    end += step
    name = "%s/%s_%d.csv" % (root_path, name_prefix, i)
    if i != (k - 1):
      data_frame[start:end].to_csv(name, header=True, index=False)
    else:
      data_frame[start:].to_csv(name, header=True, index=False)
    start += step
```
